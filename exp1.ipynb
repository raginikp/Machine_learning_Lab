```python
def train(X,y):
  for i,val in enumerate(y):
    if val=='yes':
      h=X[i].copy()
      break
      for i , val in enumerate(X):
        if y[i]=='yes':
          for z in range (len(h)):
            if val[z]!=h[z]:
              h[z]='?'
            else:
                pass
                return h
```
